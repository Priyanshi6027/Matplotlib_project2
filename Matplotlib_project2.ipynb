scatter plots

```python
import matplotlib.pyplot as plt
```

```python
day = [1,2,3,4,5]
no = [2,5,6,8,4]
color = ["r","y","g","c","b"]
plt.scatter(day,no, c = color,)
plt.title("chart", fontsize = 15)
plt.xlabel("x-axis",fontsize = 15)
plt.ylabel("y-axis",fontsize = 15)
plt.show()
```

```python
day = [1,2,3,4,5]
no = [2,5,6,8,4]
color = ["r","y","g","c","b"]
size = [2300,1220,4100,230,3515]
plt.scatter(day,no, c = color,s = size,marker ="*")
plt.title("chart", fontsize = 15)
plt.xlabel("x-axis",fontsize = 15)
plt.ylabel("y-axis",fontsize = 15)
plt.show()
```

```python
day = [1,2,3,4,5]
no = [2,5,6,8,4]
color = ["r","y","g","c","b"]
size = [2300,1220,4100,230,3515]
color =[100,333,435,765,611]
plt.scatter(day,no, c = color,s = size,marker =",", cmap = "Pastel1")
t = plt.colorbar()
t.set_label("colorbar",fontsize = 15)
plt.title("chart", fontsize = 15)
plt.xlabel("x-axis",fontsize = 15)
plt.ylabel("y-axis",fontsize = 15)
plt.show()
```

```python
day = [1,2,3,4,5]
no = [2,5,6,8,4]
no1 = [4,5,8,1,9]
color = ["r","y","g","c","b"]
size = [2300,1220,4100,230,3515]
color =[100,333,435,765,611]
plt.scatter(day,no, c = color,s = size,marker =",", cmap = "Pastel1")
plt.scatter(day,no1, c = "k",s = size,alpha = 0.5)
t = plt.colorbar()
t.set_label("colorbar",fontsize = 15)
plt.title("chart", fontsize = 15)
plt.xlabel("x-axis",fontsize = 15)
plt.ylabel("y-axis",fontsize = 15)
plt.show()
```

pie plot

```python
x = [40,10,30,20,15]
y = ["python","java","js","r","php"]
plt.pie(x,labels = y)
plt.show()
```

```python
x = [40,10,30,20,15]
y = ["python","java","js","r","php"]
c = ["k","y","r","g","c"]
ex = [0.2,0.0,0.0,0.0,0.0]
plt.pie(x,labels = y,explode = ex,colors =c)
plt.show()
```

```python
x = [40,10,30,20,15]
y = ["python","java","js","r","php"]
plt.pie(x,labels =y ,autopct = "%0.1f%%",shadow = True)
plt.show()
```

```python
x = [40,10,30,20,15]
y = ["python","java","js","r","php"]
plt.pie(x,labels =y ,autopct = "%0.1f%%",shadow = True,radius = 0.5)
plt.show()
```

```python
x = [40,10,30,20,15]
y = ["python","java","js","r","php"]
c = ["k","y","r","g","c"]
ex = [0.2,0.0,0.0,0.0,0.0]
plt.pie(x,labels = y,explode = ex,colors =c,startangle = 180)
plt.legend(loc = 1)
plt.title("pie chart")
       
plt.show()
```

```python
x = [40,10,30,20,15]
y = ["python","java","js","r","php"]
x1 = [30,23,15,20,10]
c = ["k","y","r","g","c"]

plt.pie(x,labels = y,radius = 1.5)
plt.pie(x1,radius = 1, colors =c)
plt.legend(loc = 1)
plt.title("pie chart")
       
plt.show()
```

```python
x = [40,10,30,20,15]
y = ["python","java","js","r","php"]
x1 = [30,23,15,20,10]
c = ["k","y","r","g","c"]

plt.pie(x,labels = y,radius = 1.5)
plt.pie([1], colors ="w")
plt.legend(loc = 3)
       
plt.show()
```
